```python
from sklearn.model_selection import train_test_split
```

```python
x = df1[['Age','Year', 'Genetic_Risk','Air_Pollution', 'Alcohol_Use', 'Smoking', 'Obesity_Level', 'Survival_Years']]
y = df1['Target_Severity_Score']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
predict = model.predict(x_test)
predict
```

```python
model.intercept_
```

```python
model.coef_
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score ,root_mean_squared_error,adjusted_rand_score
```

```python
mae = mean_absolute_error(y_test,predict)
mae
```

```python
mse = mean_squared_error(y_test,predict)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2 = r2_score(y_test,predict)
r2
```
